```python
# Open the file in read mode
file = open('WhatsApp Chat_Full_Format.txt', 'r', encoding="utf-8")
data = file.read()
```

```python
print(type(data)) # So, enire chat is stored in a variable in the form of string
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s' 
# pattern =   '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s?[AP]M\s-\s'
```

```python
messages = re.split(pattern, data)
messages
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
# Creating a pandas data frame
df = pd.DataFrame({'user_message':messages, 'message_date':dates})

# Convert message date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %H:%M - ')
df.rename(columns={'message_date':'date'}, inplace=True)
df.head()
```

```python
df.shape # Total 8247 messages are done
```

```python
# Separe users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_Notification')
        messages.append(entry[0])
    
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
# From df['date'] column extract year, month, day, hour and minute separetely
```

```python
df['year'] = df['date'].dt.year 
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df['day'] = df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df['mimute'] = df['date'].dt.minute
```
